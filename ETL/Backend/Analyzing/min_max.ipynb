Sure, you can create a Python class to represent the stock data and another class to represent each moment in time. The class for each moment can have the following attributes:

price: the price of the stock at that moment in time
local_min: the local minimum price up to that moment
local_max: the local maximum price up to that moment
percent_change_from_local_min: the percentage change from the local minimum
percent_change_from_local_max: the percentage change from the local maximum
The class for the stock data can have the following attributes:

data: a list of instances of the moment class, representing each moment in time
treshold: the treshold for determining local maxima and minima
You can also add methods to the stock data class for updating the local minima and maxima and for finding the datetimes for which the stock price satisfies the treshold. To confirm a local minima or maxima, you can iterate forward from that moment and look for the next local minima or maxima that satisfies the treshold. You can then return a filtered data frame with only the datetimes of the local maxima and minima.

Here is a sample implementation of the class in Python:

# Treshold PER MOment 

```python
class Moment:
    def __init__(self, price, local_min, local_max, percent_change_from_local_min, percent_change_from_local_max,date):
        self.price = price
        self.local_min = local_min
        self.local_max = local_max
        self.percent_change_from_local_min = percent_change_from_local_min
        self.percent_change_from_local_max = percent_change_from_local_max
        self.datetime = date

class StockData:
    def __init__(self, data, treshold):
        self.data = data
        self.treshold = treshold

    def update_local_min_max(self):
        for i in range(1, len(self.data)):
            if self.data[i].price < self.data[i-1].local_min:
                self.data[i].local_min = self.data[i].price
                self.data[i].percent_change_from_local_min = (self.data[i].price - self.data[i].local_min) / self.data[i].local_min * 100
            if self.data[i].price > self.data[i-1].local_max:
                self.data[i].local_max = self.data[i].price
                self.data[i].percent_change_from_local_max = (self.data[i].price - self.data[i].local_max) / self.data[i].local_max * 100

    def find_datetimes(self):
        local_minima = []
        local_maxima = []
        for i in range(1, len(self.data)):
            if self.data[i].percent_change_from_local_min < -self.treshold:
                j = i + 1
                while j < len(self.data) and self.data[j].price > self.data[i].price:
                    j += 1
                if j < len(self.data) and self.data[j].percent_change_from_local_max > self.treshold:
                    local_minima.append(self.data[i].datetime)
            if self.data[i].percent_change_from_local_max > self.treshold:
                j = i + 1
                while j < len(self.data) and self.data[j].price < self.data[i].price:
                    j += 1
                if j < len(self.data) and self.data[j].percent_change_from_local_min < -self.treshold:
                    local_maxima.append(self.data[i].datetime)
        return local_minima, local_maxima

```

```python
def find_local_min_max(prices, threshold):
    local_min = []
    local_max = []
    current_min = prices[0]
    current_max = prices[0]
    for i in range(1, len(prices)):
        price = prices[i]
        if price < current_min:
            current_min = price
        elif price >= current_min * (1 + threshold):
            local_min.append(current_min)
            current_max = price
        elif price > current_max:
            current_max = price
        elif price <= current_max * (1 - threshold):
            local_max.append(current_max)
            current_min = price
    return local_min, local_max
```
